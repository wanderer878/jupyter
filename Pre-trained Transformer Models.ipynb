```python
import torch
import transformers
from transformers import pipeline
```

```python
semantic_pipeline = pipeline('sentiment-analysis')
```

```python
sentences = [
    "I absolutely loved the movie, it was fantastic!",
     "The service was terrible and the food was disgusting.",
     "The hotel was beautiful but the staff was very rude.",
     "I went to the store to buy some groceries."
]
```

```python
for sentence in sentences:
    print(sentence)
    print(semantic_pipeline(sentence))
```

```python
specific_model = pipeline('sentiment-analysis', model= f"cardiffnlp/twitter-roberta-base-sentiment-latest")
```

```python
for sentence in sentences:
    print(sentence)
    print(specific_model(sentence))
```
