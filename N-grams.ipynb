```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt

nltk.download("punkt_tab")
from nltk.tokenize import word_tokenize
```

```python
paragraph = "Mafia II is a 2010 action-adventure game developed by 2K Czech and published by 2K. It was released on 24 August 2010 for PlayStation 3, Windows, and Xbox 360. The game is a standalone sequel to 2002's Mafia, and the second installment in the Mafia series. Set within the fictional city of Empire Bay from 1945 to 1951, the story follows Vito Scaletta, a young Sicilian-American mobster and war veteran, who becomes caught in a power struggle among the city's Mafia crime families while attempting to pay back his father's debts and secure a better lifestyle."

tokens = word_tokenize(paragraph)

tokens = [t for t in tokens if t not in [",","."]]

print(tokens)
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens,1)).value_counts())
print(unigrams)
```
