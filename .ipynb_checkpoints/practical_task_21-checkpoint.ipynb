```python
import pandas as pd
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
data = pd.read_csv("106 - tripadvisor-hotel-reviews.csv")
data.info()
data.head()
```

```python
data["Review"][0]
```

```python
#data["Review_lowercase"] = data["Review"].apply(lambda x: x.lower())

data["Review_lowercase"] = data["Review"].str.lower()
data.head()
```

```python
en_stopwords = stopwords.words('english')
en_stopwords.remove('not')
print(en_stopwords)
```

```python
data["Review_stopwords_removal"] = data["Review_lowercase"].apply(lambda x: ' '.join(word for word in \
                                                                                    x.split() if word not in \
                                                                                     en_stopwords))

data["Review_lowercase"][0]
#data.head()
```

```python
data['Review_no_punct'] = data['Review_stopwords_removal'].apply(lambda x: re.sub('[*]',' star',x))

data['Review_no_punct'][0]
```

```python
data.head()
```

```python
data['Review_no_punct'] = data['Review_stopwords_removal'].apply(lambda x: re.sub(r'([^\w\s])','',x))

data['Review_no_punct'][0]
```

```python
data.head()
```

```python
data['Review_tokenize'] = data['Review_no_punct'].apply(lambda x: word_tokenize(x))

data.head()
```

```python
Ps = PorterStemmer()

data['Review_Stemmed'] = data['Review_tokenize'].apply(lambda tokens: [Ps.stem(token) for token in tokens])
data.head()
```

```python
lemmatizer = WordNetLemmatizer()

data['Review_Lemmatized'] = data['Review_tokenize'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])

data.head()
```

```python
tokens_clean = sum(data['Review_Lemmatized'], [])
```

```python
Unigram = pd.Series(nltk.ngrams(tokens_clean, 1)).value_counts()

print(Unigram)
```

```python
bigrams = pd.Series(nltk.ngrams(tokens_clean,2)).value_counts()

print(bigrams)
```
