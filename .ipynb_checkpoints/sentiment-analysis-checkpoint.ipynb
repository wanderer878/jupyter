```python
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
sentences = [
    "I absolutely love this product, it's amazing!",
    "This is the worst experience I have ever had.",
    "This meeting is scheduled for Monday at 10am.",
    "The hotel was beautiful but the staff was rude.",
]
```

```python
print(sentences[0])
sentiment_score = TextBlob(sentences[0])
print(sentiment_score.sentiment.polarity)
```

```python
print(sentences[1])
sentiment_score = TextBlob(sentences[1])
print(sentiment_score.sentiment.polarity)
```

```python
print(sentences[2])
sentiment_score = TextBlob(sentences[2])
print(sentiment_score.sentiment.polarity)
```

```python
print(sentences[3])
sentiment_score = TextBlob(sentences[3])
print(sentiment_score.sentiment.polarity)
```

```python
vaderSentiment = SentimentIntensityAnalyzer()
```

```python
print(sentences[0])
print(vaderSentiment.polarity_scores(sentences[0]))
```

```python
print(sentences[1])
print(vaderSentiment.polarity_scores(sentences[1]))
```

```python
print(sentences[2])
print(vaderSentiment.polarity_scores(sentences[2]))
```

```python
print(sentences[3])
print(vaderSentiment.polarity_scores(sentences[3]))
```
