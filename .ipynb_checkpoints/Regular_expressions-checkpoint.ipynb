```python
import re

sentence = r"Suppose you write list. Now you want to check available methods for list\n, so you press TAB"

print(sentence)

re.search("list", sentence)
```

```python
re.sub("TAB", "TAB.", sentence) # replace krta hai
```

```python
sentence = r"The game's development began in 2011!. In order to create an accurate environment and be consistent with Tolkien's books, the developers consulted Tolkien scholars from Warner Bros, as well as Peter Jackson, director of the film trilogies The Lord of the Rings and The Hobbit. They also received assistance from Wētā Workshop, who advised on the special effects. Christian Cantamessa served as the game's lead writer, while Dan Abnett was recruited to write dialogues for the Orcs featured in the game. Combat in the game was largely influenced by the Batman: Arkham series, while Monolith had previously experimented with the nemesis system in a cancelled Batman game based on Christopher Nolan films."
sentences = sentence.split('. ') # array me tor deta hai sentence ko
print(sentences)
```

```python
pattern_to_search = r" an? " # ya to a dhondega ya an , jahan ? hga wo optional hai, isme n optional hai
arr = []

for element in sentences:
    if re.search(pattern_to_search,element):
        arr.append(element)

print(arr)
```

```python
pattern_to_search = r"^C" # ^ start of sentence me dhondega ke C hai ke nhi 
arr = []

for element in sentences:
    if re.search(pattern_to_search,element):
        arr.append(element)

print(arr)
```

```python
pattern_to_search = r"s$" # $ end of the sentence me dhondega ke s hai ke nhi
arr = []

for element in sentences:
    if re.search(pattern_to_search,element):
        arr.append(element)

print(arr)
```

```python
pattern_to_search = r"'s|Monolith" # pipe line operator , ya to sentence me 's hga ya Monolith hga
arr = []

for element in sentences:
    if re.search(pattern_to_search,element):
        arr.append(element)

print(arr)
```

```python
pattern_to_replace = r"[^\w\s]" # [^ ] ka mtlb not , \w ka mtlb word, aur \s ka mtlb whitespace 
arr = []

for element in sentences:
    modified_element =  re.sub(pattern_to_replace,"",element)
    arr.append(modified_element)

print(arr)
```
