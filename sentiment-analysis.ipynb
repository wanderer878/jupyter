```python
from textblob import TextBlob
```

```python
sentences = [
    "I absolutely love this product, it's amazing!",
    "This is the worst experience I have ever had.",
    "The meeting is scheduled for Monday at 10am.",
    "The hotel was beautiful but the staff was rude.",
]
```

```python
print(sentences[0])
sentiment_score = TextBlob(sentences[0])
print(sentiment_score.sentiment.polarity)
```
