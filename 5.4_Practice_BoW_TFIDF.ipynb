**Q2.** Create a `CountVectorizer` object and use it to fit and transform `data` in a single step. Store the result.

**Q3.** Convert the result of Q2 into a pandas DataFrame, using the vectorizer's feature names as the column headers. Print the DataFrame.

```python
print(pd.DataFrame(sparse_matrix.toarray(), columns= count_vec.get_feature_names_out()))
```

## Part 4: fit_transform vs. transform

**Q11.** Split `data` into a training set (the first 4 sentences) and a test set (the remaining sentences).

```python
data_training = data[:4]
data_test = data[4:]
print(data_training)
print(data_test)
```

**Q12.** Create a new `CountVectorizer`. Fit it on the training set only, then separately transform both the training set and the test set. Which method call do you use for the training set, and which for the test set? Why are they different?

```python
count_vec2 = CountVectorizer()
count_vec2.fit(data_training)
print(count_vec2.transform(data_training))
print('#################################################')
print(count_vec2.transform(data_test))
```

**Q13.** Convert both transformed sets into DataFrames using the vectorizer's vocabulary. Do the two DataFrames have the same number of columns? Why or why not?

```python
print(pd.DataFrame(count_vec2.transform(data_training).toarray(),columns= count_vec2.get_feature_names_out()))
print(pd.DataFrame(count_vec2.transform(data_test).toarray(),columns= count_vec2.get_feature_names_out()))
```
