## Creating a Function with a Parameter

Define a function that returns a value equal to its argument multiplied by 2.

```python
def double_func(a):
    a = a * 2;
    print(a);
    return a;

double_func(2);
```

Define a funciton that returns a float value equal to its argument divided by 2.

```python
def floater(a):
    a = a/2;
    return float(a);

floater(2)
```
