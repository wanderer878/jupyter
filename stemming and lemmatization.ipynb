```python
from nltk import PorterStemmer 
ps = PorterStemmer()
```

```python
words = ["connects" , "connecting", "connected"]

for word in words:
    print(f"{word} : {ps.stem(word)}")
```

```python
likes_tokens = ['likes', 'better', 'worse']

for word in likes_tokens:
    print(f"{word} : {ps.stem(word)}")
```

```python
tokens = ["fall", "falling", "fallen"]

for word in tokens:
    print(f"{word} : {ps.stem(word)}")
```

```python
import nltk
nltk.download("wordnet")
from nltk.stem import WordNetLemmatizer
```

```python
lem = WordNetLemmatizer()

tokens = ["fall", "falls", "falling", "fallen"]
for word in tokens:
    print(f"{word} : {lem.lemmatize(word)}")
```

```python
for word in likes_tokens:
    print(f"{word} : {lem.lemmatize(word)}")
```
