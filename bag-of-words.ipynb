```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data = [
  "Tomb Raider 2 takes Lara Croft on a fast-paced adventure across the globe.",
  "The game opens with Lara chasing a mysterious dagger through Venice.",
  "Lara must navigate deadly traps and ancient ruins in Tibet.",
  "In China, she discovers a secret temple hidden beneath a monastery.",
  "The snow-covered cliffs and icy caverns are some of the game's most memorable locations.",
  "Lara faces ruthless enemies and skilled mercenaries at every turn.",
  "The game's soundtrack builds tension during each daring escape.",
  "Lara uses her climbing, swimming, and shooting skills to survive.",
  "The story centers on an artifact that could change history forever.",
  "Tomb Raider 2 blends exploration, puzzle-solving, and action in a classic way."
]
```

```python
count_vec = CountVectorizer()
```

```python
# creating a sparse matrix , this will only remember  where values are 1 so saves up 
# memory
sparse_matrix = count_vec.fit_transform(data) 
print(sparse_matrix)
```

```python
# now it will populate all the 0 values , sparse will give only 1, saves up efficiency
bag_of_words = pd.DataFrame(sparse_matrix.toarray(),columns= count_vec.get_feature_names_out())
print(bag_of_words)
```
