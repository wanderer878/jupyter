```python
from nltk import PorterStemmer 
ps = PorterStemmer()
```

```python
words = ["connects" , "connecting", "connected"]

for word in words:
    print(f"{word} : {ps.stem(word)}")
```

```python
likes_tokens = ['likes', 'better', 'worse']

for word in likes_tokens:
    print(f"{word} : {ps.stem(word)}")
```
