```python
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")
```

```python
en_stopwords = stopwords.words('english')

print(en_stopwords)
```

```python
en_stopwords.remove("but")
en_stopwords.append("code")

my_sentence = "because the code exists, but the data file is missing."

short_sentence = " ".join([word for word in my_sentence.split() if word not in en_stopwords])

print(short_sentence)
```
